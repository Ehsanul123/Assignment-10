```python
#5:
import os
import shutil
def moveFileType(folder):
    
    for root, dirs, files in os.walk(folder):
        
        for file in files:
            
            if file.endswith('.pdf'):
                
                image_path=os.path.join(root,file)
                print('location: ',image_path)  
                shutil.copy(image_path, 'E:\Maxima')

```

```python
moveFileType('C:')
```
